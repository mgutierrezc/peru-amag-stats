```python
import os
import boto3, botocore
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
session = boto3.session.Session()
client = session.client('s3',
                        config=botocore.config.Config(s3={'addressing_style': 'virtual'}),
                        region_name='fra1',
                        endpoint_url='https://fra1.digitaloceanspaces.com',
                        aws_access_key_id=os.getenv('SPACES_KEY'),
                        aws_secret_access_key=os.getenv('SPACES_SECRET'))
```

```python
response = client.list_objects(Bucket='peru-judges-data')
```

```python
for obj in response['Contents']:
    print(obj['Key'])
```

```python
client.download_file('peru-judges-data',
                     'peru/pdf_10_1/ACEVEDO DIEZ CECILIA DEL PILAR/._Resolucion_CINCO_2021-10-28 12_59_37.754.pdf',
                     '../Data/Judges/._Resolucion_CINCO_2021-10-28 12_59_37.754.pdf')
```
